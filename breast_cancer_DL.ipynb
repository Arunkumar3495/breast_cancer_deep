```python
X = df.drop(columns = 'target',axis=1)
Y = df['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_std = scaler.fit_transform(X_train)
X_test_std  = scaler.transform(X_test)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_test.shape)
print(Y_train.shape)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
```

```python
from tensorflow import keras
```

```python
# setting up the layer of neutral network
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20,activation = 'relu'),
                          keras.layers.Dense(2,activation = 'sigmoid')

])
```

```python
# compiling the neural networks
model.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
# training the neural network
histroy = model.fit(X_train,Y_train,validation_split=0.1,epochs = 10)
```

```python
# training the neural network
Histroy = model.fit(X_train_std,Y_train,validation_split=0.1,epochs = 10)
```

```python
plt.plot(histroy.histroy['accuracy'])
plt.plot(histroy.histroy['val_accuracy'])

plt.title("model accuracy")
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.legend("training data","validation data",loc="lower right")
```

```python
# accuracy on the test data

loss,accuracy = model.evaluate(X_test_std,Y_test)
print(accuracy)
```
